# Resume Screening Proxy Bias Experiment

## Purpose
This project investigates whether machine learning models used for resume screening can learn and reproduce age-related disparities **without explicit age features**, by relying on correlated proxy variables (e.g., experience, titles, technology signals, and optionally graduation year).

## Core Research Question
Can a resume-screening model produce systematically different outcomes for older candidates even when age is not provided, due to proxy feature leakage?

## High-Level Approach
1. Generate a synthetic resume dataset with realistic correlations between latent age group and observable resume features.
2. Define a simulated screening outcome (e.g., callback) with adjustable bias strength.
3. Train baseline models using only non-protected features.
4. Evaluate performance and fairness metrics across age groups.
5. Run ablation experiments to measure how removing strong proxies (e.g., graduation year) changes disparities.

## Notebooks
- `01_data_generation.ipynb`: synthetic dataset creation and persistence
- `02_model_training.ipynb`: model training and baseline performance
- `03_fairness_evaluation.ipynb`: fairness metrics + subgroup performance analysis
- `04_ablation_graduation_year.ipynb`: controlled ablation study focused on graduation year
